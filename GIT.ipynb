```python
data_panen = {
    'lokasi1': {
        'nama_lokasi': 'Kebun A',
        'hasil_panen': {
            'padi': 1200,
            'jagung': 800,
            'kedelai': 500
        }
    },
    'lokasi2': {
        'nama_lokasi': 'Kebun B',
        'hasil_panen': {
            'padi': 1500,
            'jagung': 900,
            'kedelai': 450
        }
    },
    'lokasi3': {
        'nama_lokasi': 'Kebun C',
        'hasil_panen': {
            'padi': 1100,
            'jagung': 750,
            'kedelai': 600
        }
    },
    'lokasi4': {
        'nama_lokasi': 'Kebun D',
        'hasil_panen': {
            'padi': 1300,
            'jagung': 850,
            'kedelai': 550
        }
    },
    'lokasi5': {
        'nama_lokasi': 'Kebun E',
        'hasil_panen': {
            'padi': 1400,
            'jagung': 950,
            'kedelai': 480
        }
    }
}
```

```python
import pandas as pd
df = pd.DataFrame(data_panen)

df
```

```python
jumlah_jagung = df.loc['hasil_panen', 'lokasi2']['jagung']

print(f"Jumlah hasil panen jagung dari lokasi2 adalah: {jumlah_jagung}")
```

```python
nama_lokasi_3 = df['lokasi3']['nama_lokasi']

print(f"Nama lokasi dari lokasi3 adalah: {nama_lokasi_3}")
```

```python
hasil_panen_dict = df.loc['hasil_panen']

padi_per_lokasi = hasil_panen_dict.apply(lambda x: x['padi'])

kedelai_per_lokasi = hasil_panen_dict.apply(lambda x: x['kedelai'])

print("Hasil Panen Padi per Lokasi (padi_per_lokasi)")
print(padi_per_lokasi)

print("\nHasil Panen Kedelai per Lokasi (kedelai_per_lokasi)")
print(kedelai_per_lokasi)
```

```python
print("Hasil Analisis Kondisi Lokasi:")

for lokasi_key in df.columns:
    
    nama_lokasi = df[lokasi_key]['nama_lokasi']
    
    hasil = df[lokasi_key]['hasil_panen']
    padi = hasil['padi']
    jagung = hasil['jagung']
    
    if padi > 1300 or jagung > 800:
        status = "memerlukan perhatian khusus"
        keterangan = f"(Padi: {padi}, Jagung: {jagung})"
        print(f"ðŸ”´ {nama_lokasi} ({lokasi_key}) {status} {keterangan}")
    else:
        status = "dalam kondisi baik"
        keterangan = f"(Padi: {padi}, Jagung: {jagung})"
        print(f"ðŸŸ¢ {nama_lokasi} ({lokasi_key}) {status}. {keterangan}")
```
